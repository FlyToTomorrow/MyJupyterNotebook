```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

fig_format = '.svg'
# fig_format = '.tif'
fig_dvi = 1600
LW = 2
FS = 14
FS_smaller = 12
FS_ticklabel = 14
FS_ticklabel_small = 12
FS_label = 14
FS_legend = 13

# plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=FS)
```

```python
t_start, t_end = -1e-6, 1e-6
num = 3e4
t_step = (t_end - t_start) / num
f0 = 1 / 100e-9
amp = 0.3
offset = 0
t = np.linspace(t_start, t_end, int(num))
wav1 = amp * signal.sawtooth(2 * np.pi * f0 * t, 0.5) + offset
wav2 = np.exp(t / 1e-6 - 1)

plt.plot(t * 1e6, wav1, 'b', label='f(t)')
plt.plot(t * 1e6, wav2, 'g', label='g(t)')
plt.plot(t * 1e6, wav1 * wav2, 'r', label='_nolegend_')
plt.fill_between(t * 1e6, wav1 * wav2, facecolor='r', alpha=0.5, label='f(t) * g(t)')
plt.legend(loc='upper left', ncol=1)
plt.xlim(-1, 1)
plt.show()

result = sum(wav1 * wav2) * t_step
print('result = {0}'.format(result))
```
