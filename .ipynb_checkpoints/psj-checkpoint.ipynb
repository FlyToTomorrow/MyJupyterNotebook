# Psj

## Setup

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

fig_format = '.svg'
# fig_format = '.tif'
fig_dvi = 1600
LW = 2
FS = 14
FS_smaller = 12
FS_ticklabel = 14
FS_ticklabel_small = 12
FS_label = 14
FS_legend = 13

# plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=FS)
```

## Test 1 

```python
t_start, t_end = -1e-6, 1e-6
num = 3e4
t_step = (t_end - t_start) / num
f0 = 1 / 100e-9
amp = 0.3
offset = 0
t = np.linspace(t_start, t_end, int(num))
wav1 = amp * signal.sawtooth(2 * np.pi * f0 * t, 0.5) + offset
wav2 = np.exp(t / 1e-6 - 1)

plt.plot(t * 1e6, wav1, 'b', label='f(t)')
plt.plot(t * 1e6, wav2, 'g', label='g(t)')
plt.plot(t * 1e6, wav1 * wav2, 'r', label='_nolegend_')
plt.fill_between(t * 1e6, wav1 * wav2, facecolor='r', alpha=0.5, label='f(t) * g(t)')
plt.legend(loc='upper left', ncol=1)
plt.xlim(-1, 1)
plt.show()

result = sum(wav1 * wav2) * t_step
print('result = {0}'.format(result))
```

## Test 2

```python
t_start, t_end = -10e-6, 1e-6
num = 3e4
t_step = (t_end - t_start) / num

f0 = 1 / 100e-9
amp = 0.3
offset = 0
t = np.linspace(t_start, t_end, int(num))
wav1 = amp * signal.sawtooth(2 * np.pi * f0 * t, 0.5) + offset
wav2 = np.exp(t / 1e-6 - 1)

result = sum(wav1 * wav2) * t_step
print('result = {0}'.format(result))
```

## Test 3 (20180601)

```python
## Create waveform 1

# Waveform 1, part 1
# parameters
t1_start, t1_end = -10e-6, 1e-6+100e-9/4
num = 3e4
t_step = (t1_end - t1_start) / num
f0 = 1 / 100e-9
amp = 0.3
offset = 0
# create waveform
t1 = np.linspace(t1_start, t1_end, int(num))
wav1_part1 = amp * signal.sawtooth(2 * np.pi * f0 * t1, 0.5) + offset

# Waveform 1, part 2
# parameters
t2_start, t2_end = 1e-6+100e-9/4, 1.5e-6+100e-9/4
f0 = 1 / 500e-9
amp = 1.5/2
offset = amp
t2 = np.arange(t2_start, t2_end+t_step, t_step)
t2_shift = t2-(1e-6+100e-9/4)
wav1_part2 = amp * signal.sawtooth(2 * np.pi * f0 * t2_shift, 0.5) + offset
# create waveform
t = np.append(t1, t2)
wav1 = np.append(wav1_part1, wav1_part2)


## Create waveform 1
wav2 = np.exp(t / 1e-6 - 1)

## Plot waveform
plt.figure(figsize=(12, 8))
plt.plot(t * 1e6, wav1, 'b', label='f(t)')
plt.plot(t * 1e6, wav2, 'g', label='g(t)')
plt.plot(t * 1e6, wav1 * wav2, 'r', label='_nolegend_')
plt.fill_between(t * 1e6, wav1 * wav2, facecolor='r', alpha=0.5, label='f(t) * g(t)')
plt.legend(loc='upper left', ncol=1)
plt.xlim(-1, 2)
plt.show()

## Calculate the integral
result = sum(wav1 * wav2) * t_step
print('result = {0}'.format(result))
```
